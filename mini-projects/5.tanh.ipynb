```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def tanh(x):
  return (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))

def der_tanh(x):
  return 1 - tanh(x)*tanh(x)
```

```python
x=np.linspace(-10,10,100)
y=tanh(x)
dy=der_tanh(x)
```

```python
plt.plot(x,y,x,dy)
plt.title('Tanh & Derivative')
plt.legend(['tanh','der_tanh'])
plt.grid()
plt.show()
```
