<div style="font-family:BNazanin" dir="rtl">
    <h2>1- تابع فعال ساز RuLU</h2>
    <p>معادله تابع فعال ساز به صورت روبرو است: f(x) = max(0,x) </br>
    که نشان می دهد که اگر ورودی x کمتر از صفر باشد، خروجی تابع صفر خواهد بود و اگر ورودی داده شده به تابع relu بزرگتر از صفر باشد، خروجی تابع همان ورودی خواهد بود.
    </p>
</div>

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-50,50)
y = np.maximum(0,x)

plt.plot(x,y)
plt.grid()
plt.show()
```

<div style="font-family:BNazanin" dir="rtl">
    <h2>2- مشکل ReLU در حال مرگ</h2>
    <p>با توجه به فرمول تابع فعال سازی ReLU، تمامی مقادیر کمتر از صفر پس از عبور از تابع ReLU صفر خواهند شد. اگر شبکه عصبی مقادیر منفی زیادی در مجموعه داده داشته باشد، همه آنها پس از عبور از تابع فعال سازی ReLU به صفر تبدیل می شوند و صفرهایی در خروجی وجود دارد که نورون ها را غیرفعال می کند. این نوع سناریو در شبکه های عصبی مشکل ReLu در حال مرگ نامیده می شود.</p>
    <p>برای حل مشکلات ReLU در حال مرگ در شبکه های عصبی، انواع دیگری از تابع فعال سازی ReLU توسعه یافته است که با مشکل ReLU در حال مرگ یا نورون های در حال مرگ مواجه نمی شوند.</p>
</div>

<div style="font-family:BNazanin" dir="rtl">
    <h2>3- Leaky ReLU (LReLU)</h2>
    <p>اولین و ساده ترین نوع تابع فعال سازی ReLU، ReLU نشتی است. تنها کمی تغییر در فرمول تابع فعال سازی عادی ReLu وجود دارد که مشکل مرگ ReLU را حل می کند. فرمول تابع فعال‌سازی Leaky ReLU به این صورت است: </br> f(x) = max(0.1x , x)</p>
</div>

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-50,50)
y = np.maximum(0.1*x,x)

plt.plot(x,y)
plt.grid()
plt.show()
```

<div style="font-family:BNazanin" dir="rtl">
    <h2>4- Parametric ReLU (PReLU)</h2>
    <p>ReLU پارامتری مشابه تابع فعال‌سازی ReLU نشتی است، اما در اینجا به‌جای داشتن یک مقدار از پیش تعیین‌شده برای ضرب مقادیر ورودی در ۰.۰۱، از پارامتر «a» در ReLU پارامتریک استفاده می‌کنیم. مقدار پارامتر "a" توسط خود شبکه عصبی تعیین می شود.
فرمول پارامتریک ReLU به صورت ذیل است: </br> f(x) = max(ax,x)</p>
</div>

```python
import numpy as np
import matplotlib.pyplot as plt

a = np.random.randn()
x = np.linspace(-50,50)
y = np.maximum(a*x,x)

plt.plot(x,y)
plt.grid()
plt.show()
```

<div style="font-family:BNazanin" dir="rtl">
    <h2>5- Exponential Linear Unit (ELU)</h2>
    <p> گونه‌ای از فعال‌سازی ReLU است که برخی مشکلات را برطرف می‌کند و در مقایسه با تابع فعال‌سازی معمولی ReLU مزایایی را به همراه دارد. فرمول ELU به صورت ذیل است: </br> f(x) = x for x >=0 </br> f(x) = a(pow(e,x)-1) for other  </p>
    </br>
    <p>تابع فعال‌سازی ELU از یک منحنی log استفاده می‌کند و به همین دلیل، برخلاف ReLU نشتی و پارامتریک، خط مستقیمی برای مقادیر منفی نمی‌دهد.

توابع فعال سازی ELU همچنین مشکل ReLU در حال مرگ را حل می کند و همچنین منحنی ELU برای مقادیر منفی با قدر بالا به آرامی صاف می شود در حالی که Relu معمولی فوراً صاف می شود.</p>
</div>

```python
import numpy as np
import matplotlib.pyplot as plt

a = np.random.randn()
x = np.linspace(-50,50)
y = []
for i in x:
    if i >= 0:
        y.append(i)
    else:
        y.append(a*(np.exp(i)-1))

plt.plot(x,y)
plt.grid()
plt.show()
```

<div style="font-family:BNazanin" dir="rtl">
    <h2>6- Scaled Exponential Linear Unit (SELU)</h2>
    <p>f(x) = landa*x for x >=0 </br> f(x) = landa*a(pow(e,x)-1) for other </br>
    معادله بیان می کند که اگر مقدار ورودی x بزرگتر از صفر باشد، مقدار خروجی x ضرب در لامبدا λ می شود و اگر مقدار ورودی x کمتر یا مساوی صفر باشد، آلفا با نمایی مقدار x ضرب می شود. منهای مقدار آلفا، و سپس در مقدار لامبدا ضرب می کنیم.
</p>
</div>

```python
import numpy as np
import matplotlib.pyplot as plt

a = np.random.randn()
landa = np.random.randn()
x = np.linspace(-50,50)
y = []
for i in x:
    if i > 0:
        y.append(landa*i)
    else:
        y.append(landa*a*(np.exp(i)-1))

plt.plot(x,y)
plt.grid()
plt.show()
```
