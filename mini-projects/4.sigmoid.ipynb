```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))

def der_sigmoid(x):
  return sigmoid(x)*(1-sigmoid(x))
```

```python
x=np.linspace(-10,10,100)
y=sigmoid(x)
dy=der_sigmoid(x)
```

```python
plt.plot(x,y,x,dy)
plt.title('Sigmoid & Derivative')
plt.legend(['sigmoid','der_sigmoid'])
plt.grid()
plt.show()
```
