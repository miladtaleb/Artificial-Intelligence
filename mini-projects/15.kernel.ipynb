```python
from skimage import data
import matplotlib.pyplot as plt
import cv2

image = data.astronaut()
```

```python
image.shape
```

```python
plt.figure(figsize=(5,5))
plt.imshow(image)
```

```python
import numpy as np
kernel = np.array([[0,
-1,
0
],[-1,
5,
-1],[0,
-1,
0]])
```

```python
convolved_img=cv2.filter2D(image,-1,kernel)
```

```python
fig,ax=plt.subplots(1,2,figsize=(16,32))
ax[0].imshow(image)
ax[0].axis('off')
ax[0].set_title('original')

ax[1].imshow(convolved_img)
ax[1].axis('off')
ax[1].set_title('sharpen')
```

```python
kernel=np.array([[0.0625,
0.125,
0.0625],[0.125,
0.25,
0.125],[0.0625,
0.125,
0.0625]])
```

```python
convolved_img=cv2.filter2D(image,-1,kernel)
fig,ax=plt.subplots(1,2,figsize=(16,32))
ax[0].imshow(image)
ax[0].axis('off')
ax[0].set_title('original')

ax[1].imshow(convolved_img)
ax[1].axis('off')
ax[1].set_title('blurred')
```
