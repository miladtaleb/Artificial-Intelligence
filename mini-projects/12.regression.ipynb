```python
import numpy as np
import tensorflow as tf
```

```python
x=np.linspace(-2*np.pi,2*np.pi,10000).reshape(-1,1)
y=np.sin(x)
```

```python
# train test -> shuffle = True
from sklearn.model_selection import train_test_split


X_train,X_test,y_train,y_test=train_test_split(x,y,shuffle=True)



# batch_size = 16,32
# epochs = 1000

# loss? optimizer? metrics?


# model? layers? activation='sigmoid'  input,hidden,output

# eval: mse? r2?
```

```python
model = tf.keras.models.Sequential()

model.add(tf.keras.layers.Dense(units=30,activation='sigmoid'))
model.add(tf.keras.layers.Dense(units=15,activation='sigmoid'))
model.add(tf.keras.layers.Dense(units=1))
#model.add(tf.keras.layers.Dense(units=1,activation='linear'))
```

```python
model.compile(loss='mean_squared_error',optimizer='adam',metrics=['mean_squared_error'])
```

```python
history=model.fit(X_train,y_train,epochs=500,batch_size=16)
```

```python
model.summary()
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred[:10]
```

```python
from sklearn.metrics import mean_squared_error,r2_score

print('mse',mean_squared_error(y_test,y_pred))
print('r2',r2_score(y_test,y_pred))
```
